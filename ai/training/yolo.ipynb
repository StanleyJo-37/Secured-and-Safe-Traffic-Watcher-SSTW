```python
from ultralytics import YOLO
import torch
```

```python
torch.cuda.empty_cache()
```

```python
yolo_model = YOLO('yolov8n.pt')
```

```python
results = yolo_model.train(
    data="a9_dataset.yaml",
    epochs=100,
    patience=10,
    imgsz=640,
    batch=32,
    device=0,
    project="yolo_sstw",
    name="yolov8n_standard"
)
```

```python
import os
yolo_model.export(format='onnx', simplify=True)
yolo_model.export(format='onnx', simplify=True, half=True)
```
